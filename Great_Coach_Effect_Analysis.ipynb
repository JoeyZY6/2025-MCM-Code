```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# 假设 test_data 已经加载
# 示例：test_data = pd.read_csv('summerOly_athletes.csv')

# 过滤出所需的数据
selected_projects = {
    'SWE': 'Volleyball',
    'ESP': 'Swimming',
    'BUL': 'Table Tennis',
}

filtered_data = test_data[test_data.apply(lambda row: row['NOC'] in selected_projects and row['Sport'] == selected_projects[row['NOC']], axis=1)]

# 删除不必要的列（如果存在）
columns_to_drop = ['Name', 'City', 'Sex', 'Team', 'Event']
existing_columns = [col for col in columns_to_drop if col in filtered_data.columns]
filtered_data = filtered_data.drop(columns=existing_columns)

print("Filtered data shape:", filtered_data.shape)

# 数据预处理：确保所有特征列都是数值类型，并填充缺失值
filtered_data = filtered_data.apply(pd.to_numeric, errors='coerce').fillna(0)

# 特征工程
# 假设 Medal 是我们的目标变量，其他列为特征
features = [col for col in filtered_data.columns if col not in ['Medal']]
X = filtered_data[features]
y = filtered_data['Medal']

# 添加常数项
X_const = sm.add_constant(X)

print("Features used for training:", X_const.columns.tolist())

# 检查是否有无效值
if X_const.isnull().values.any():
    print("Invalid values found in X_const")
    print(X_const[X_const.isnull().any(axis=1)])
else:
    print("No invalid values found in X_const")

# 训练线性回归模型
model_lr = sm.OLS(y, X_const).fit()
print(model_lr.summary())

# 准备测试数据
def prepare_test_data(test_data, train_features, project):
    # 创建一个空的DataFrame，列名为训练时的特征列
    mapped_df = pd.DataFrame(columns=train_features)
    
    for feature in train_features:
        if feature in test_data.columns:
            mapped_df[feature] = test_data[feature].apply(pd.to_numeric, errors='coerce').fillna(0)
        else:
            # 如果特征不存在，填充默认值（例如0）
            mapped_df[feature] = 0
    
    # 设置特定项目的特征列为1
    if project in train_features:
        mapped_df[project] = 1
    
    # 确保常数项存在
    if 'const' not in mapped_df.columns:
        mapped_df = sm.add_constant(mapped_df)
    
    # 确保列顺序一致
    mapped_df = mapped_df[train_features]
    
    # 检查是否有无效值
    if mapped_df.isnull().values.any():
        print("Invalid values found in mapped_df")
        print(mapped_df[mapped_df.isnull().any(axis=1)])
        return None
    
    return mapped_df

# 预测每个选定的弱势项目的当前成绩
current_performance = {}

for country, project in selected_projects.items():
    # 获取特定国家和项目的训练数据
    project_data = filtered_data[(filtered_data['NOC'] == country) & (filtered_data['Sport'] == project)]
    
    if len(project_data) < 2:  # 确保有足够的数据点
        print(f"Insufficient data for {country} - {project}: {len(project_data)} samples")
        continue
    
    # 使用与训练时相同的特征列
    X = prepare_test_data(project_data, X_const.columns.tolist(), project)
    
    if X is None:
        continue
    
    # 打印X的形状和列名，确保与训练数据一致
    print(f"Shape of X for {country} - {project}: {X.shape}")
    print(f"Columns of X for {country} - {project}: {X.columns.tolist()}")
    
    # 检查是否有无效值
    if X.isnull().values.any():
        print(f"Invalid values found in X for {country} - {project}")
        continue
    
    # 预测当前成绩
    predictions_lr = model_lr.predict(X)
    
    if predictions_lr.isnull().values.any():
        print(f"NaN predictions for {country} - {project}")
        continue
    
    current_performance[(country, project)] = predictions_lr.mean()

print("Current performance of selected projects:")
for (country, project), perf in current_performance.items():
    print(f"{country} - {project}: {perf:.2f}")

# 估计引入“名帅”后的成绩提升
regression_coefficient = 0.1146
ate = 0.4479
lift = 1.9731

def estimate_improvement_with_great_coach(current_performance):
    estimated_improvements = {}
    
    for (country, project), perf in current_performance.items():
        # 使用回归系数估计提升
        improvement_by_coeff = regression_coefficient
        
        # 使用ATE估计提升
        improvement_by_ate = ate
        
        # 使用Lift估计提升
        improvement_by_lift = lift * perf
        
        # 综合三种方法的估计值
        improvement = (improvement_by_coeff + improvement_by_ate + improvement_by_lift) / 3
        
        estimated_improvements[(country, project)] = improvement
    
    return estimated_improvements

estimated_improvements = estimate_improvement_with_great_coach(current_performance)

print("Estimated improvements with a great coach:")
for (country, project), imp in estimated_improvements.items():
    print(f"{country} - {project}: {imp:.2f}")

# 详细分析是否应该招募‘名帅’
print("Detailed analysis for recruiting a great coach:")
for (country, project), current_perf in current_performance.items():
    est_imp = estimated_improvements[(country, project)]
    print(f"Country: {country}, Sport: {project}")
    print(f"Current Performance: {current_perf:.2f}")
    print(f"Estimated Improvement with Great Coach: {est_imp:.2f}")
    print("-" * 40)
```

```python
import pandas as pd
import statsmodels.api as sm



# 删除不必要的列（如果存在）
columns_to_drop = ['Name', 'City', 'Sex', 'Team', 'Event']
existing_columns = [col for col in columns_to_drop if col in test_data.columns]
test_data = test_data.drop(columns=existing_columns)

# 计算每个国家的总奖牌数
country_medal_counts = test_data.groupby('NOC')['Medal'].sum().sort_values(ascending=False)

# 选择前N个强势国家（例如，前20个国家）
top_countries = country_medal_counts.head(20).index

# 对于每个强势国家，找出其弱势项目
weak_projects_per_country = {}

for country in top_countries:
    # 获取该国家的所有数据
    country_data = test_data[test_data['NOC'] == country]
    
    # 计算每个项目的总奖牌数
    project_medal_counts = country_data.groupby('Sport')['Medal'].sum().sort_values()
    
    # 选择奖牌数最少的三个项目作为弱势项目
    weak_projects = project_medal_counts.head(3).index
    
    weak_projects_per_country[country] = weak_projects.tolist()

print("Weak projects per strong country:")
for country, projects in weak_projects_per_country.items():
    print(f"{country}: {projects}")

# 获取训练时使用的特征列名（注意这里应包含常数项）
train_features = X_train_const.columns.tolist()

def prepare_test_data(test_data, train_features):
    # 创建一个空的DataFrame，列名为训练时的特征列
    mapped_df = pd.DataFrame(columns=train_features)
    
    for feature in train_features:
        if feature in test_data.columns:
            mapped_df[feature] = test_data[feature]
        else:
            # 如果特征不存在，填充默认值（例如0）
            mapped_df[feature] = 0
    
    # 确保常数项存在
    if 'const' not in mapped_df.columns:
        mapped_df = sm.add_constant(mapped_df)
    
    # 确保列顺序一致
    mapped_df = mapped_df[train_features]
    
    # 检查是否有无效值
    if mapped_df.isnull().values.any():
        print("Invalid values found in mapped_df")
        print(mapped_df[mapped_df.isnull().any(axis=1)])
        return None
    
    return mapped_df

# 预测每个选定的弱势项目的当前成绩
current_performance = {}

def predict_current_performance(country, sport, train_features, model_lr):
    # 获取特定国家和项目的训练数据
    project_data = test_data[(test_data['NOC'] == country) & (test_data['Sport'] == sport)]
    
    if len(project_data) < 2:  # 确保有足够的数据点
        print(f"Insufficient data for {country} - {sport}: {len(project_data)} samples")
        return None
    
    # 使用与训练时相同的特征列
    X = prepare_test_data(project_data, train_features)
    
    if X is None:
        return None
    
    # 打印X的形状和列名，确保与训练数据一致
    print(f"Shape of X for {country} - {sport}: {X.shape}")
    print(f"Columns of X for {country} - {sport}: {X.columns.tolist()}")
    
    # 检查是否有无效值
    if X.isnull().values.any():
        print(f"Invalid values found in X for {country} - {sport}")
        return None
    
    # 预测当前成绩
    predictions_lr = model_lr.predict(X)
    
    if predictions_lr.isnull().values.any():
        print(f"NaN predictions for {country} - {sport}")
        return None
    
    return predictions_lr.mean()

for country, projects in weak_projects_per_country.items():
    current_performance[country] = {}
    for project in projects:
        performance = predict_current_performance(country, project, train_features, model_lr)
        if performance is not None:
            current_performance[country][project] = performance
        else:
            print(f"No valid prediction for {country} - {project}")

print("Current performance of weak projects:")
for country, projects in current_performance.items():
    for project, perf in projects.items():
        print(f"{country} - {project}: {perf:.2f}")

# 估计引入“名帅”后的成绩提升
regression_coefficient = 0.1146
ate = 0.4479
lift = 1.9731

def estimate_improvement_with_great_coach(current_performance):
    estimated_improvements = {}
    
    for country, projects in current_performance.items():
        estimated_improvements[country] = {}
        for project, perf in projects.items():
            # 使用回归系数估计提升
            improvement_by_coeff = regression_coefficient
            
            # 使用ATE估计提升
            improvement_by_ate = ate
            
            # 使用Lift估计提升
            improvement_by_lift = lift * perf
            
            # 综合三种方法的估计值
            improvement = (improvement_by_coeff + improvement_by_ate + improvement_by_lift) / 3
            
            estimated_improvements[country][project] = improvement
    
    return estimated_improvements

estimated_improvements = estimate_improvement_with_great_coach(current_performance)

print("Estimated improvements with a great coach:")
for country, projects in estimated_improvements.items():
    for project, imp in projects.items():
        print(f"{country} - {project}: {imp:.2f}")

# 选择三个国家的三个体育项目进行详细分析
print("Selected weak projects for analysis:")
for i, (country, projects) in enumerate(weak_projects_per_country.items()):
    if i >= 3:
        break
    for j, project in enumerate(projects):
        if j >= 1:
            break
        
        # 检查是否在current_performance和estimated_improvements中有对应的键
        if country in current_performance and project in current_performance[country]:
            current_perf = current_performance[country][project]
        else:
            current_perf = "No data"
        
        if country in estimated_improvements and project in estimated_improvements[country]:
            est_imp = estimated_improvements[country][project]
        else:
            est_imp = "No data"
        
        print(f"Country: {country}, Sport: {project}")
        print(f"Current Performance: {current_perf}")
        print(f"Estimated Improvement with Great Coach: {est_imp}")
        print("-" * 40)
```
