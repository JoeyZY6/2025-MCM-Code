```python
# Complxified Model
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Dropout, Concatenate
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
import tensorflow as tf

# 假设 athlete_host_data 包含 'NOC', 'Year', 'Event', 'Medal' 列
# 特征提取与聚合
grouped_data = athlete_host_data.groupby(['NOC', 'Year']).agg({'Medal': 'sum'}).reset_index()
pivot_data = grouped_data.pivot(index='Year', columns='NOC', values='Medal').fillna(0)

# 创建一个标志位列，表示是否为东道主国家
host_country = 'USA'
pivot_data['Host'] = (pivot_data.index >= 2028) & (pivot_data[host_country] == pivot_data.index)
pivot_data['Host'] = pivot_data['Host'].astype(int)

# 时间序列构建
time_series_data = pivot_data.values
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(time_series_data[:, :-1])  # 不标准化 Host 列

def create_dataset(dataset, host_flags, time_step):
    X, Y, H = [], [], []
    for i in range(len(dataset) - time_step):
        a = dataset[i:(i + time_step), :]
        h = host_flags[i:(i + time_step)]
        X.append(a)
        H.append(h)
        Y.append(dataset[i + time_step, :])
    return np.array(X), np.array(H), np.array(Y)

time_step = 20  # 使用过去12年的数据来预测下一年
X, H, Y = create_dataset(scaled_data, pivot_data['Host'].values[:-1], time_step)

# 确保 X 和 Y 的长度一致
assert len(X) == len(Y), "Length of X and Y should be the same"

# 设置样本权重
weights = np.ones(len(X))
weights[pivot_data.index[:len(X)] >= 2012] = 2  # 2012年及以后的数据权重为20，其余为1

# 划分训练集和测试集
train_size = int(len(X) * 0.8)
test_size = len(X) - train_size
X_train, X_test = X[:train_size], X[train_size:]
H_train, H_test = H[:train_size], H[train_size:]
Y_train, Y_test = Y[:train_size], Y[train_size:]
weights_train = weights[:train_size]

# 定义分位数损失函数
def quantile_loss(q, y_true, y_pred):
    e = y_true - y_pred
    return tf.keras.backend.mean(tf.keras.backend.maximum(q*e, (q-1)*e), axis=-1)

# 构建LSTM模型并指定分位数
def build_model(input_shape, units=150, dropout=0.2, output_units=X_train.shape[2], q=0.5):
    inputs = Input(shape=input_shape)
    host_inputs = Input(shape=(input_shape[0], 1))  # 主办国标志位输入
    
    x = LSTM(units=units, return_sequences=True, kernel_regularizer=tf.keras.regularizers.l2(0.01))(inputs)
    x = Dropout(dropout)(x)
    x = LSTM(units=units, return_sequences=True, kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
    x = Dropout(dropout)(x)
    x = LSTM(units=units, kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
    x = Dropout(dropout)(x)
    
    # 处理主办国标志位
    host_x = LSTM(units=units, return_sequences=True, kernel_regularizer=tf.keras.regularizers.l2(0.01))(host_inputs)
    host_x = Dropout(dropout)(host_x)
    host_x = LSTM(units=units, return_sequences=True, kernel_regularizer=tf.keras.regularizers.l2(0.01))(host_x)
    host_x = Dropout(dropout)(host_x)
    host_x = LSTM(units=units, kernel_regularizer=tf.keras.regularizers.l2(0.01))(host_x)
    host_x = Dropout(dropout)(host_x)
    
    # 将两个分支的结果合并
    combined = Concatenate()([x, host_x])
    outputs = Dense(output_units, activation='relu')(combined)  # 使用ReLU激活函数确保输出非负
    
    model = Model(inputs=[inputs, host_inputs], outputs=outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
                  loss=lambda y_true, y_pred: quantile_loss(q, y_true, y_pred))
    
    return model

# 训练两个模型分别用于下界和上界的预测
quantiles = [0.8, 0.9]  # 分别代表下界和上界对应的分位数
models = {}

# 回调函数
early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)

for q in quantiles:
    print(f"Training model for quantile {q}")
    model = build_model((X_train.shape[1], X_train.shape[2]), q=q)
    model.fit([X_train, H_train], Y_train, epochs=300, batch_size=32, sample_weight=weights_train, 
              validation_split=0.2, callbacks=[early_stopping, reduce_lr], verbose=1)
    models[q] = model

# 预测2028年的下界和上界
last_known_data = scaled_data[-time_step:, :]
last_known_hosts = pivot_data['Host'].values[-time_step:].reshape(-1, 1)
future_pred_lower = models[quantiles[0]].predict([last_known_data.reshape((1, last_known_data.shape[0], last_known_data.shape[1])), 
                                                  last_known_hosts.reshape((1, last_known_hosts.shape[0], 1))])
future_pred_upper = models[quantiles[1]].predict([last_known_data.reshape((1, last_known_data.shape[0], last_known_data.shape[1])), 
                                                  last_known_hosts.reshape((1, last_known_hosts.shape[0], 1))])

# 调整形状以适应 inverse_transform
future_pred_lower = future_pred_lower.reshape(future_pred_lower.shape[0], -1)
future_pred_upper = future_pred_upper.reshape(future_pred_upper.shape[0], -1)

# 反向转换以获取原始尺度上的预测值
future_pred_lower = scaler.inverse_transform(future_pred_lower)
future_pred_upper = scaler.inverse_transform(future_pred_upper)

# 创建DataFrame并丢弃包含NA的预测结果
future_pred_df_lower = pd.DataFrame(future_pred_lower, columns=pivot_data.columns[:-1])
future_pred_df_upper = pd.DataFrame(future_pred_upper, columns=pivot_data.columns[:-1])
future_pred_df_lower.dropna(axis=1, inplace=True)
future_pred_df_upper.dropna(axis=1, inplace=True)

# 确保下界不大于上界
predictions_2028 = {}
for country in future_pred_df_lower.columns:
    lower = max(0, future_pred_df_lower[country].values[0])  # 确保下界非负
    upper = max(lower, future_pred_df_upper[country].values[0])  # 确保上界不小于下界
    predictions_2028[country] = (lower, upper)

# 显示结果
for country, (lower, upper) in predictions_2028.items():
    print(f"{country}: [{lower:.2f}, {upper:.2f}] medals")
```

```python
# Test with USA host
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Dropout, Concatenate
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
import tensorflow as tf

# 假设 athlete_host_data 包含 'NOC', 'Year', 'Event', 'Medal' 列
# 特征提取与聚合
grouped_data = athlete_host_data.groupby(['NOC', 'Year']).agg({'Medal': 'sum'}).reset_index()
pivot_data = grouped_data.pivot(index='Year', columns='NOC', values='Medal').fillna(0)

# 创建一个标志位列，表示是否为东道主国家
host_country = 'USA'
pivot_data['Host'] = (pivot_data.index >= 2028) & (pivot_data[host_country] == pivot_data.index)
pivot_data['Host'] = pivot_data['Host'].astype(int)

# 时间序列构建
time_series_data = pivot_data.values
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(time_series_data[:, :-1])  # 不标准化 Host 列

def create_dataset(dataset, host_flags, time_step):
    X, Y, H = [], [], []
    for i in range(len(dataset) - time_step):
        a = dataset[i:(i + time_step), :]
        h = host_flags[i:(i + time_step)]
        X.append(a)
        H.append(h)
        Y.append(dataset[i + time_step, :])
    return np.array(X), np.array(H), np.array(Y)

time_step = 12  # 使用过去12年的数据来预测下一年
X, H, Y = create_dataset(scaled_data, pivot_data['Host'].values[:-1], time_step)

# 确保 X 和 Y 的长度一致
assert len(X) == len(Y), "Length of X and Y should be the same"

# 设置样本权重
weights = np.ones(len(X))
weights[pivot_data.index[:len(X)] >= 2012] = 2  # 2012年及以后的数据权重为3，其余为1

# 划分训练集和测试集
train_size = int(len(X) * 0.8)
test_size = len(X) - train_size
X_train, X_test = X[:train_size], X[train_size:]
H_train, H_test = H[:train_size], H[train_size:]
Y_train, Y_test = Y[:train_size], Y[train_size:]
weights_train = weights[:train_size]

# 定义分位数损失函数
def quantile_loss(q, y_true, y_pred):
    e = y_true - y_pred
    return tf.keras.backend.mean(tf.keras.backend.maximum(q*e, (q-1)*e), axis=-1)

# 构建LSTM模型并指定分位数
def build_model(input_shape, units=100, dropout=0.2, output_units=X_train.shape[2], q=0.5):
    inputs = Input(shape=input_shape)
    host_inputs = Input(shape=(input_shape[0], 1))  # 主办国标志位输入
    
    x = LSTM(units=units, return_sequences=True, kernel_regularizer=tf.keras.regularizers.l2(0.01))(inputs)
    x = Dropout(dropout)(x)
    x = LSTM(units=units, kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
    x = Dropout(dropout)(x)
    
    # 处理主办国标志位
    host_x = LSTM(units=units, return_sequences=True, kernel_regularizer=tf.keras.regularizers.l2(0.01))(host_inputs)
    host_x = Dropout(dropout)(host_x)
    host_x = LSTM(units=units, kernel_regularizer=tf.keras.regularizers.l2(0.01))(host_x)
    host_x = Dropout(dropout)(host_x)
    
    # 将两个分支的结果合并
    combined = Concatenate()([x, host_x])
    outputs = Dense(output_units, activation='relu')(combined)  # 使用ReLU激活函数确保输出非负
    
    model = Model(inputs=[inputs, host_inputs], outputs=outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
                  loss=lambda y_true, y_pred: quantile_loss(q, y_true, y_pred))
    
    return model

# 训练两个模型分别用于下界和上界的预测
quantiles = [0.8, 0.85]  # 分别代表下界和上界对应的分位数
models = {}

# 回调函数
early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)

for q in quantiles:
    print(f"Training model for quantile {q}")
    model = build_model((X_train.shape[1], X_train.shape[2]), q=q)
    model.fit([X_train, H_train], Y_train, epochs=300, batch_size=32, sample_weight=weights_train, 
              validation_split=0.2, callbacks=[early_stopping, reduce_lr], verbose=1)
    models[q] = model

# 预测2028年的下界和上界
last_known_data = scaled_data[-time_step:, :]
last_known_hosts = pivot_data['Host'].values[-time_step:].reshape(-1, 1)
future_pred_lower = models[quantiles[0]].predict([last_known_data.reshape((1, last_known_data.shape[0], last_known_data.shape[1])), 
                                                  last_known_hosts.reshape((1, last_known_hosts.shape[0], 1))])
future_pred_upper = models[quantiles[1]].predict([last_known_data.reshape((1, last_known_data.shape[0], last_known_data.shape[1])), 
                                                  last_known_hosts.reshape((1, last_known_hosts.shape[0], 1))])

# 调整形状以适应 inverse_transform
future_pred_lower = future_pred_lower.reshape(future_pred_lower.shape[0], -1)
future_pred_upper = future_pred_upper.reshape(future_pred_upper.shape[0], -1)

# 反向转换以获取原始尺度上的预测值
future_pred_lower = scaler.inverse_transform(future_pred_lower)
future_pred_upper = scaler.inverse_transform(future_pred_upper)

# 创建DataFrame并丢弃包含NA的预测结果
future_pred_df_lower = pd.DataFrame(future_pred_lower, columns=pivot_data.columns[:-1])
future_pred_df_upper = pd.DataFrame(future_pred_upper, columns=pivot_data.columns[:-1])
future_pred_df_lower.dropna(axis=1, inplace=True)
future_pred_df_upper.dropna(axis=1, inplace=True)

# 确保下界不大于上界
predictions_2028 = {}
for country in future_pred_df_lower.columns:
    lower = max(0, future_pred_df_lower[country].values[0])  # 确保下界非负
    upper = max(lower, future_pred_df_upper[country].values[0])  # 确保上界不小于下界
    predictions_2028[country] = (lower, upper)

# 显示结果
for country, (lower, upper) in predictions_2028.items():
    print(f"{country}: [{lower:.2f}, {upper:.2f}] medals")
```
